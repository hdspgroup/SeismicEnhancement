```python
import gempy as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pyvista
import gempy_viewer as gpv
import gempy as plot_2d
```

```python
geo_model: gp.data.GeoModel = gp.create_geomodel(
    project_name='Modelo_1',
    extent=[0, 10000, 0, 2000, -2000, 500],
    resolution=[20, 20, 20],  # * Here we define the resolution of the voxels
    importer_helper=gp.data.ImporterHelper(
        path_to_orientations= "OneDrive/Documentos/TESIS/SSJmodelo/orientaciones_sinclinal.txt",
        path_to_surface_points= "OneDrive/Documentos/TESIS/SSJmodelo/Sinclinal_Sinu_superficies.txt",
    )
)
```

```python
geo_model.surface_points
```

```python
geo_model.orientations
```

```python
geo_model.structural_frame
```

```python
gp.map_stack_to_surfaces(
    gempy_model=geo_model,
    mapping_object={
        "Fault_Series": ('Falla'),
        "Strat_Series": ('NgQs', 'Ngc','Pgc')
    }
)

geo_model.structural_frame  # Display the resulting structural frame

# Set identified faults in the structural frame
gp.set_is_fault(
    frame=geo_model.structural_frame,
    fault_groups=['Fault_Series']
)
```
